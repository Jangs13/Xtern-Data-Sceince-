### For Finding Shops

```python
#searching type 
search_string = 'shops'
distance = 500
#to store the data 
place_list = []

#getting the info using places_nearby()
response = gmaps.places_nearby(
    location = (39.768809067428464, -86.15800717599967),
    keyword = search_string,
    radius = distance
)

place_list.extend(response.get('results'))
next_page_token = response.get('next_page_token')

#running in the loop to save response in place_list
while next_page_token:
    time.sleep(2)
    response = gmaps.places_nearby(
    location = (39.76727775800655, -86.17888450031133),
    keyword = search_string,
    radius = distance,
    page_token = next_page_token
    )
    place_list.extend(response.get('results'))
    next_page_token = response.get('next_page_token')
    
```

```python
#saving the data in csv file
df = pd.DataFrame(place_list)
df['url'] = 'https://www.google.com/maps/place/?q=place_id:' + df['place_id']
df.to_csv('Data_Shops.csv', index = False)
df= df.drop(['business_status','geometry','icon', 'icon_background_color', 'icon_mask_base_uri'
        ,'opening_hours', 'photos', 'plus_code', 'reference', 'scope', 'price_level', 'place_id', 'user_ratings_total'], axis =1)
df.to_csv('Data_Shops.csv', index = False)
```

### Combine all dataset into one csv sheet


```python
import glob
import pandas as pd

# specifying the path to csv files
path = "C:\\Users\\athar\\Xtern"

# csv files in the path
file_list = glob.glob(path + "/*.csv")

# list of csv files we want to merge.
# pd.read_csv(file_path) reads the csv
# data into pandas dataframe.
csv_list = []

for file in file_list:
    csv_list.append(pd.read_csv(file))

# create a new dataframe to store the
# merged csv file.
csv_merged = pd.DataFrame()

for csv_file in csv_list:

    # appends the data into the csv_merged
    # dataframe.
    csv_merged = csv_merged.append(
    csv_file, ignore_index=True)

# exports the dataframe into csv file with
# specified name.
csv_merged.to_csv('places_data.csv', index=False)
```
